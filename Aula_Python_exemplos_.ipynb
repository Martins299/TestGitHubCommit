Atividade 2: Sinais de Comparação
Objetivo: Praticar o uso de operadores de comparação.

Instruções:

Crie um script Python que compare dois números e exiba os resultados das seguintes comparações:

Maior que.

Menor que.

Igual a.

Diferente de.

Maior ou igual a.

Menor ou igual a.

Exemplo de Código:

```python
x = 10
y = 5
print("x > y:", x > y)
print("x < y:", x < y)
print("x == y:", x == y)
print("x != y:", x != y)
print("x >= y:", x >= y)
print("x <= y:", x <= y)
```

Atividade 5: Manipulação de Strings como Listas
Objetivo: Praticar a manipulação de strings como listas.

Instruções:

Crie um script Python que defina uma string e exiba:
O primeiro caractere.
O último caractere.
Uma fatia da string (por exemplo, do segundo ao quinto caractere).
O comprimento da string.
Exemplo de Código:

```python
texto = "Ciência da Computação"
print("Primeiro caractere:", texto[0])
print("Último caractere:", texto[-1])
print("Fatia da string:", texto[0:10])
print("Comprimento da string:", len(texto))
```
